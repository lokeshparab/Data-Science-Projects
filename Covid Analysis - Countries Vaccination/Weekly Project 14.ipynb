# COVID Analysis - Sorting Countries

## Import libraries

```python
import pandas as pd
import numpy as np
```

## Loading dataset

```python
df = pd.read_csv('country_vaccination_preprocessed.csv')
df.head()
```

```python
# getting sum of total vaccinations group by country
data = df.groupby('country').sum()['total_vaccinations']
```

```python
# dataframe presentation of output
def dataframe(df):
    df_ = pd.DataFrame()
    df_['country'] = df.index
    df_['total_vaccinations'] = df.values
    return df_
```

##  1.Which Country is having least total Vaccinations?

```python
nw_data = data.sort_values().head()
dataframe(nw_data)
```

## 2.Which Country is having the most total Vaccinations?

```python
nw_data = data.sort_values( ascending = False).head()
dataframe(nw_data)
```
