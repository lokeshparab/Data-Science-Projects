# COVID Analysis - Country Specific

## Import libraries

```python
import pandas as pd
import numpy as np
```

## Loading dataset

```python
df = pd.read_csv('country_vaccination_preprocessed.csv')
df.head()
```

```python
# converting float to int type of column total vaccinations
df['total_vaccinations'] = df['total_vaccinations'].apply( lambda x : int(x))
```

```python
def dataframe(df):
    df_ = pd.DataFrame()
    df_['month'] = df.index
    df_['total_vaccinations'] = df.values
    return df_
```

## Find Monthly number of total vaccination from Jan to Dec in 2021

```python
# collection of data for year 2021
df_2021 = df[df['year'] == 2021]

# Getting sum of total vaccinations group by months
df_monthly = df_2021.groupby('month_name').sum()['total_vaccinations']

# final output
dataframe(df_monthly)
```
