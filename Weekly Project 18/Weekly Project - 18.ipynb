## Import Libraries

```python
import pandas as pd
import numpy as  np

from matplotlib import pyplot as plt
```

## Loading Dataset

```python
df = pd.read_csv('Product.csv')
df
```

```python
df.info()
```

## Name and Orders Analysis

```python
data = df.sort_values(by='Orders',ascending = False)
orders = data['Orders']
names  = data['Name']

plt.rc('font', **{'size':30})
plt.figure(figsize=(100,50))
plt.pie(orders,labels = names,autopct = '%1.2f%%')
plt.legend()
plt.title('Name and Orders Analysis', fontsize = 60)
plt.show()
```

## Conclusion: 
### UBS-C Charging Cable and Lighting Charging are in demanding and most purchase by people whereas LG Driver and LG Washing Machine have most less demands in Market

## Name and Max Quantity Per User(Column) Analysis

```python
data = df.sort_values(by='Max_quantity_per_user',ascending = False).head()

x    = data['Name']
y    = data['Max_quantity_per_user']

plt.figure(figsize=(25,10))
plt.rc('font', **{'size':20})
color = ['red','orange','yellow','green','blue' , 'violet', 'pink',"brown", 'black', 'magenta', 'grey', 'skyblue', 'maroon','lightpink','darkblue', 'lightgreen', 'lightblue', ]

ax = plt.bar(x,y , width = 0.6, color = color)

for k, i in enumerate(ax):
    
    bar_width = i.get_width() 
    x_cord    = i.xy[0] + bar_width / 2
    y_cord    = i.get_height()
    txt       = str(y_cord)
    #print(y_cord)
    plt.annotate(
        text      = txt,
        xy        = (x_cord , y_cord),
        xytext    = ( 0, 4),
        textcoords="offset points"
        
                )

plt.xlabel('Products')
plt.ylabel('Price')
plt.title('Name and Max Quantity Per User(Column) Analysis (Top 5)')
plt.show()
```

## Conclusion

### AAA Batteries(4-pack) is found to be most max quantity per user among the other products
