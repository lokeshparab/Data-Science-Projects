## Import Libraries

```python
import pandas as pd
import numpy as  np

from matplotlib import pyplot as plt
```

## Loading Dataset

```python
df = pd.read_csv('Product.csv')
df
```

```python
df.info()
```

## Name and Orders Analysis

```python
data = df.sort_values(by='Orders',ascending = False)
orders = data['Orders']
names  = data['Name']

def format(p,orders):
    value = int(p/100 * orders.sum())
    return f'{round(p,2)}% ({value})'

plt.rc('font', **{'size':32})
plt.figure(figsize=(100,70))
plt.pie(orders,labels = names,autopct = lambda p : format(p,orders))
plt.legend()
plt.title('Name and Orders Analysis', fontsize = 60)
plt.show()
```

## Conclusion: 
### UBS-C Charging Cable and Lighting Charging are in demanding and most purchase by people whereas LG Driver and LG Washing Machine have most less demands in Market

## Name and Max Quantity Per User(Column) Analysis

```python
data = df.sort_values(by='Max_quantity_per_user',ascending = False)

names    = data['Name']
max_     = data['Max_quantity_per_user']

def format(p,max_):
    value    = p/100 * max_.sum()
    
    decimal  = value - int(value) 
    
    value   += 1 if decimal > 0.5 else 0
    return f'{int(value)}'

plt.rc('font', **{'size':30})
plt.figure(figsize=(40,40))
plt.pie(max_,labels = names,autopct = lambda p : format(p,max_))
plt.legend()

plt.title('Name and Max Quantity Per User(Column) Analysis')
plt.show()
```

## Conclusion

### AAA Batteries(4-pack) , AA Batteries(4-pack) and USB-C Cable is found to be  max quantity per user more than 5 hence availabity of Product need to done to such product

## Name and Amount Analysis

```python
data = df.sort_values(by='Amount',ascending = False)

names    = data['Name']
Amount     = data['Amount']

def format(p,Amount):
    value  = p/100 * Amount.sum()
    
    value += 1 if float(str(6.99)[1:]) > 0.5 else 0
    return f'{round(p,2)}% ({int(value)})'

plt.rc('font', **{'size':30})
plt.figure(figsize=(60,60))
plt.pie(Amount,labels = names,autopct = lambda p : format(p,Amount))
plt.legend()

plt.title('Name and Amount Analysis')
plt.show()
```

## Conclusion

### MacBookPro Laptop and Iphone have earn highest amount in the sales report market 

## Create V2 of Product.csv with Additional Information

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.info()
```

```python
data = df.groupby('product')
price          = data.mean()['price']

quantity_total = data.sum()[['quantity', 'total']]

size           = data.size()

max_quantity   = data.max()['quantity']
```

```python
product = pd.DataFrame()

product['Name']                = size.index
product['Price']               = price.values
product['Orders']              = size.values
product[['Quantity','Amount']] = quantity_total.values
product['Max Quantity']        = max_quantity.values

product['Quantity'] = product['Quantity'].apply(lambda x: int(x))
product.head()
```

```python
product.to_csv('ProductV2.csv')
```
