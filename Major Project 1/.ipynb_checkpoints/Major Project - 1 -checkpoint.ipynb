# Write a program to create a dictionary with product id as the key and product details as values for inventory management system

* **Detail Structure of Product in Dictionary**
```python
 { key --> product_id : values --> [ product_name, category, price, quantity, discount ] }
```

## Importing Libraries

````python
import json as js
import numpy as np
import pandas as pd
````

## **Read and Write function from Json Files**

````python
# reads records.json file and return dictionary 

def read_file():
    
    # Opening records.json file in read mode
    fw = open('records.json','r')
    
    # Loading json file into dictionary
    r = js.loads(fw.read())
    
    # Closing the file
    fw.close()
    
    # Return dictionary
    return r
````

````python
#Write records.json file for given dictionaty input 

def write_file(var):
    
    # Opening records.json file in write mode
    fw = open('records.json','w')
    
    # Dumping dictionary into json file  
    fw.write(js.dumps(var))
    
    # Closing the file
    fw.close()
````
