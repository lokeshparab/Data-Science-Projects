# Import Library

```python
import requests as rq
import json     as js

from bs4 import BeautifulSoup as BS
```

# Github Profile Scraper for all the Followers

```python
# Taking Github user as input 
while True:
    user_id = input("Enter Github user id :")
    res  = rq.get(f'https://github.com/{user_id}?tab=followers')
    if res.ok : break
    else      : print("Invalid user id try again") 

# follower dictionary data
follower = {}

# Souping the content and fetching follower tags
soup = BS(res.content,'html.parser')
followers = soup.find('turbo-frame').find_all(
    'div', 
    class_ = 'd-table table-fixed col-12 width-full py-4 border-bottom color-border-muted')

# Visiting for each followers
for person in followers:
    
    #Scraping image link, user_id & user_link
    profile  = person.find('img')
    img_link = profile.get('src').split('?')[0]
    usr_id  = profile.get('alt')[1:]
    user_link = f'https://github.com/{user_id}'
    
    # Scraping person's name
    name     = person.find('span').text.strip()
    
    # Scraping person's bio
    bio      = person.find('div', class_ = 'color-fg-muted text-small mb-2')
    bio      = bio.text.strip() if bio   else 'None'
    
    # Scraping organistion & location
    org_loc  = person.find('p',class_='color-fg-muted text-small mb-0')
    data     = []
    
    # if organistion & location is present 
    if org_loc:
        
         # Scraping person's organisation
        org  = org_loc.find('span')
        org  = org.text.strip() if org   else 'None'
        
        # Scraping person's location
        loc  = org_loc.text.strip()
        loc  = loc if type(org) is float else loc    .split('\n')
        loc  = loc if type(loc) is str   else loc[-1].strip()     if len(loc) != 1 else 'None'
        
        data = [org,loc]    
    
    # default null value for organistion & location
    else:
        loc      = 'None'
        org      = 'None'
        data = [org,loc]
        
    # fetching all data in dictionary
    follower[usr_id] = {
        'image_link' : img_link,
        'user_name'  : name,
        'bio'        : bio,
        'organisation': data[0],
        'location'    : data[1],
        'user_link'   : user_link
    }

with open(f'{user_id}.json','w') as fd:
    fd.write(js.dumps(follower))
    
print(f'Output available in {user_id}.json file')
```
