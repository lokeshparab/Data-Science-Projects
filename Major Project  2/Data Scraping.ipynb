## Multiple page web scraping

```python
import requests as rq
from bs4 import BeautifulSoup as BS
from tqdm import tqdm
import pandas as pd

flag = False
new_link = '/s?k=books'

lst = []
while new_link:
    
    if not type(new_link) == str:
        new_link = new_link.get('href')
        
    link = 'https://www.amazon.in' + new_link
    #print(link)
    res     = rq.get(link)
    while not res.ok:
        res     = rq.get(link)

    soup    = BS( res.content, 'html.parser')
    content = soup.find('div', class_ = 's-main-slot s-result-list s-search-results sg-row')
    new_link = content.find('a',class_ = 's-pagination-item s-pagination-next s-pagination-button s-pagination-separator')
    books   = content.find_all('div', class_ = 'a-section a-spacing-small a-spacing-top-small')

    for book in books:

        # Getting book name , author, and release date
        name_author_release    = book.find('div', class_ = 'a-section a-spacing-none puis-padding-right-small s-title-instructions-style')
        name                   = name_author_release.find('span').text
        book_link              = name_author_release.find('a').get('href')
        edition_author_release = name_author_release.find('div' )
        
        if edition_author_release is None:
            edition_author_release = [ float('nan') for i in range(3)]
        
        else:
            edition_author_release = edition_author_release.text.split('|')
            
            if len(edition_author_release) == 2:
                edition_author_release = [float('nan')] + edition_author_release

        # getting  ratings and rating count
        ratings = book.find_all('div', class_ = 'a-section a-spacing-none a-spacing-top-micro')

        if len(ratings) == 1:
            ratings       = float('nan')
            rating_counts = float('nan')
        else:
            ratings       = ratings[0].text
            rating_counts = ratings.split(')')[-1]
            ratings       = ratings[:3]


       # gettting price of book
        price = book.find('span', class_ = 'a-price a-text-price')
        
        if price is None:
            price = book.find('span', class_ = 'a-price-whole')
            
            if price is None:
                price = float('nan')
            
            else:
                price = price.text
            
        else:
            price = price.find('span').text[1:]
            
        
        # combing all scraped in list and added in final lst
        data = [name, book_link] + edition_author_release + [ratings, rating_counts, price]
        
        lst.append(data)

columns = ['name','link','edition','author','release','ratings','rating_counts','price']
df = pd.DataFrame( lst , columns = columns)
df.head()
```

```python
df.to_csv('AmazonBooks.csv',index = False)
```

```python
# scraping top 100 books of each genres

import pandas   as pd
import requests as rq

from bs4  import BeautifulSoup as BS
from tqdm import tqdm

genres = pd.read_csv('Amazongenres.csv')

book_df = []

for genre, link in tqdm(genres.values):
    
    while link:
        
        if type(link) != str:
            link = 'https://www.amazon.in' + link.get('href')  
            
        res   = rq.get(link)
        
        while not res.ok : res = rq.get(link)

        soup  = BS(res.content, 'html.parser')

        soup  = soup.find('div', class_ = 'a-cardui _cDEzb_card_1L-Yx')
        
        books = soup.find_all('div', class_ = 'a-cardui _cDEzb_grid-cell_1uMOS expandableGrid p13n-grid-content')
        
        for book in books:
            
            rank         = book.find('div').text[1:]
            book_link    = book.find('a'  ).get('href')
            image_link   = book.find('img').get('src' )
            name         = book.find('img').get('alt' )
            
            author       = book.find('div', class_ = 'a-row a-size-small' )
            reviews      = book.find('div', class_ = 'a-icon-row'         )
            
            book_link    = 'https://www.amazon.in' + book_link
            
            try       : author_name  = author.text
            except    : author_name  = float('nan')
            
            if reviews:
                ratings_and_counts   = [ i.text       for i in reviews.find_all('span') ]
                
                reviews_link         = 'https://www.amazon.in' + reviews.find('a').get('href')
                
                reviews              = ratings_and_counts      + [ reviews_link ]
                
            else:
                reviews              = [ float('nan') for i in range(3)                 ]
            
            try       : author_link  = 'https://www.amazon.in' +  author.find('a').get('href') 
                
            except    : author_link  = float('nan')
                 
            
            data         = [ name, book_link, genre, image_link, rank, author_name, author_link]
            data        += reviews
            
            book_df.append(data)
        
        
        link  = soup.find('li', class_ = 'a-last').find('a')

columns = ['name','link','genres','image','rank','author','author_link','ratings','counts','reviews']
df = pd.DataFrame(book_df, columns = columns)
df.to_csv('Amazonbookgenres.csv', index=False)
```

```python
# getting paperback link from all books

import pandas   as pd
import requests as rq
import numpy    as np

from bs4  import BeautifulSoup as BS
from tqdm import tqdm

def paperback(link):
    
    res          = rq.get(link)
        
    while not res.ok: res = rq.get(link)

    soup         = BS( res.content, 'html.parser' )

    editions     = soup.find('ul' , class_ = 'a-unordered-list a-nostyle a-button-list a-horizontal')
    
    link         = None
    
    for edition in editions.find_all('li'):
        
        if 'paperback' in edition.text.lower():
            
            link = 'https://www.amazon.in' + edition.find('a').get('href')
    
    return link

df               = pd.read_csv('Amazonbookgenres.csv')

links            = []

for link in tqdm(df['link']):
    if 'ebook' in link:  
        links.append(paperback(link))
    else:
        links.append(link)

df_              = df[['name','link']]
df_['paperlink'] = links

df_.to_csv('Amazonbookgenres_paperlink.csv', index= False)
```
