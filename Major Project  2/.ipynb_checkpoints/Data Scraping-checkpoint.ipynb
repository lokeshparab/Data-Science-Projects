# Student Data scraping

```python
import pandas as pd

df      = pd.read_excel('College_Students.xlsx')


data    = list( df.iloc[ 3   : 82  ,1 : 4].values )
data   += list( df.iloc[ 88  : 161 ,1 : 4].values )
data   += list( df.iloc[ 166 : 214 ,1 : 4].values )
data   += list( df.iloc[ 220 : 322 ,1 : 4].values )

data   += list( df.iloc[ 327 : 374 ,1 : 4].values )
data   += list( df.iloc[ 379 : 456 ,1 : 4].values )
data   += list( df.iloc[ 461 : 527 ,1 : 4].values )
data   += list( df.iloc[ 532 : 541 ,1 : 4].values )

data   += list( df.iloc[ 546 : 591 ,1 : 4].values )
data   += list( df.iloc[ 596 : 632 ,1 : 4].values )
data   += list( df.iloc[ 637 : 697 ,1 : 4].values )
data   += list( df.iloc[ 702 : 721 ,1 : 4].values )


df      = pd.read_excel('College_Students.xlsx', sheet_name = 1)


data   += list( df.iloc[ 3   : 139 ,1 : 4] .values )
data   += list( df.iloc[ 144 : 178 ,1 : 4].values )
data   += list( df.iloc[ 183 : 303 ,1 : 4].values )
data   += list( df.iloc[ 702 : 721 ,1 : 4].values )

columns =       df.iloc[ 2         ,1 : 4].values

df      = pd.DataFrame(data, columns = columns)


def grad_year( branch ):
    
    year      = { 'TE': 23 , 'BE': 22 , 'SE': 24 }
    data      = branch.split(' ')[0]
    return      2000 + year[data]
    
def degree( branch ):
    degree_name      = { 'comp' : '( Computer Engineering )'   ,
                         'it'   : '( Information Technology )' , 
                         'elex' : '( Electronics & Electrical Engineering )',
                         'extc' : '( Electronics & Telecommunication Engineering )' }
    
    name             =  branch.split(' ')[1].lower()
    return              'BE ' + degree_name[name]
    
    
df['Graduation year'] = df['Branch'].apply(grad_year)
df['degree'         ] = df['Branch'].apply(degree   )

del df['Branch']

students = {}
for student in df.values:
    
    students[ student[0] ] = {
        'name'            : student[1],
        'graduation year' : student[2],
        'degree'          : student[3],
        'issued'          : {}
    }

import json as js
import os

# write dictionary data into json file 
def write_file(var, name = 'book' ):
    
    # writing extension of file as .json
    name += 's.json'
    
    # Opening json file in write mode
    fw    = open( name ,'w')

    # Dumping dictionary into json file  
    fw.write(js.dumps(var))

    # Closing the file
    fw.close()

write_file(students,'student')

df.head()
```

## Multiple page web scraping

```python
import requests as rq
from bs4 import BeautifulSoup as BS
from tqdm import tqdm
import pandas as pd

flag = False
new_link = '/s?k=books'

lst = []
while new_link:
    
    if not type(new_link) == str:
        new_link = new_link.get('href')
        
    link = 'https://www.amazon.in' + new_link
    #print(link)
    res     = rq.get(link)
    while not res.ok:
        res     = rq.get(link)

    soup    = BS( res.content, 'html.parser')
    content = soup.find('div', class_ = 's-main-slot s-result-list s-search-results sg-row')
    new_link = content.find('a',class_ = 's-pagination-item s-pagination-next s-pagination-button s-pagination-separator')
    books   = content.find_all('div', class_ = 'a-section a-spacing-small a-spacing-top-small')

    for book in books:

        # Getting book name , author, and release date
        name_author_release    = book.find('div', class_ = 'a-section a-spacing-none puis-padding-right-small s-title-instructions-style')
        name                   = name_author_release.find('span').text
        book_link              = name_author_release.find('a').get('href')
        edition_author_release = name_author_release.find('div' )
        
        if edition_author_release is None:
            edition_author_release = [ float('nan') for i in range(3)]
        
        else:
            edition_author_release = edition_author_release.text.split('|')
            
            if len(edition_author_release) == 2:
                edition_author_release = [float('nan')] + edition_author_release

        # getting  ratings and rating count
        ratings = book.find_all('div', class_ = 'a-section a-spacing-none a-spacing-top-micro')

        if len(ratings) == 1:
            ratings       = float('nan')
            rating_counts = float('nan')
        else:
            ratings       = ratings[0].text
            rating_counts = ratings.split(')')[-1]
            ratings       = ratings[:3]


       # gettting price of book
        price = book.find('span', class_ = 'a-price a-text-price')
        
        if price is None:
            price = book.find('span', class_ = 'a-price-whole')
            
            if price is None:
                price = float('nan')
            
            else:
                price = price.text
            
        else:
            price = price.find('span').text[1:]
            
        
        # combing all scraped in list and added in final lst
        data = [name, book_link] + edition_author_release + [ratings, rating_counts, price]
        
        lst.append(data)

columns = ['name','link','edition','author','release','ratings','rating_counts','price']
df = pd.DataFrame( lst , columns = columns)
df.head()
```

```python
df.to_csv('AmazonBooks.csv',index = False)
```

```python
# getting genres for booking
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup as BS
from tqdm import tqdm

df            = pd.read_csv('AmazonBooks.csv')

genres = []
for link in tqdm(df['link']):
    
    link      = 'https://www.amazon.in' + link
    
    res       = rq.get(link)
    
    while not res.ok:
        res   = rq.get(link)
    
    book_page = BS( res.content, 'html.parser')
    
    genre     = book_page.find('span',class_ = 'cat-link')
    
    genre     = genre.text if genre else float('nan')
    
    genres   += [genre]
```

```python
df['genres'] = genres
df.to_csv('AmazonBooks.csv',index = False)
```

## Preprocessing of Amazon books

```python
import pandas as pd
df = pd.read_csv('AmazonBooks.csv')
df.head()
```

```python
df['price'        ] = df['price'        ].apply( lambda x : float(x.replace(',','')) if type(x) == str else x )
df['rating_counts'] = df['rating_counts'].apply( lambda x :   int(x.replace(',','')) if type(x) == str else x )
df['author'       ] = df['author'       ].apply( lambda x :       x.strip()[3:]      if type(x) == str else x )
df['genres'       ] = df['genres'       ].apply( lambda x :       x[3:]              if type(x) == str else x )
df['link'         ] = df['link'         ].apply( lambda x : 'https://www.amazon.in' + x                       )

df.head()
```
