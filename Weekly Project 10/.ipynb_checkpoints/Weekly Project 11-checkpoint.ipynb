#  COVID Analysis - Data Preprocessing

# Import libraries

```python
import pandas as pd
import numpy as np
```

# Loading dataset

```python
df = pd.read_csv('country_vaccination_preprocessed.csv')
df.head()
```

```python
df.info()
```

```python
date = df['date']
date
```

# Adding Year Column in the dataset from Date Column.

```python
def year(date):
    return date.split('-')[0]

df['year'] = date.apply(year)
```

# Adding Month Column in the dataset from Date Column (1-12)

```python
def month(date):
    return date.split('-')[1]

df['month'] = date.apply(month)
df['month'].unique()
```

# Add Day Column in the dataset from Date Column

```python
def day(date):
    return date.split('-')[-1]

df['day'] = date.apply(day)
df['day'].unique()
```

```python
df.head()
```

```python
df.to_csv('country_vaccination_preprocessed.csv',index = False )
```
