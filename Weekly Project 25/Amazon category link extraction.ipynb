# Amazon Sales Scraper

## Import Library

```python
import pandas as pd
import requests as rq

from bs4                          import BeautifulSoup as BS
from selenium                     import webdriver     as wd
from selenium.webdriver.common.by import By
```

```python
link = "https://www.amazon.in/"


driver = wd.Chrome('webdriver\chromedriver')

driver.get(link)

anchor_tag = driver.find_element(By.ID,"nav-hamburger-menu")
anchor_tag.click()

soup = BS(driver.page_source, 'html.parser')
```

```python
category = soup.find('div', id ='hmenu-content')
```

```python
dct = {
    'maincategory'   :[],
    'subcategory'    :[],
    'category'       :[],
    'link'           :[],
     }

for sr,i in enumerate(category.find_all('ul', class_='hmenu hmenu-translateX-right')[7:-1]):
    sub = None
    for j in i.find_all('li')[1:]:
        
        link_ref = j.find('a')
        
        if link_ref: 
            cat      = j.text.strip()
            print('*',cat)
            link_ref = link_ref.get('href')
            link_ref = link[:-1] + link_ref if 'gp' in link_ref else link_ref
            print(link_ref)
            
            dct['maincategory'].append(sr)
            dct['subcategory'].append(sub)
            dct['category'].append(cat)
            dct['link'].append(link_ref)
        
        elif j.text.strip() == '': continue
        else:
            
            sub = j.text.strip()
            print('-->',sub)
    print('-'*40)
```

```python
data = pd.DataFrame(dct)
```

```python
data['link'] = data['link'].apply(lambda x : link[:-1] + x if x[0]=='/' else x)
```
